## 创建Series数据类型

```python
import numpy as np
pd.Series(np.arange(1,7))
```

```python
index=['a','b','c','d']
s1=pd.Series(np.random.randn(4),index=index)
s1
```

```python
index=['e','f']
s2=pd.Series(np.random.randn(2),index=index)
s1.append(s2)#拼接
```

```python
#删除指定的索引
s1.drop('a')  
```

```python
s1   #没有记忆性
```

```python
s1['a']
```

```python
#访问前三个数据
s1[:3]
```

```python
dates=pd.date_range('today',periods=6) #从今天开始产生6天
num_arr=np.random.randn(6,4)#6X4的正态分布矩阵
columns=['A','B','C','D']#每列的索引
df1=pd.DataFrame(num_arr,index=dates,columns=columns)  #不指定index将会以0-6为索引
df1
```
